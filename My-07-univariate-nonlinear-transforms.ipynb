```python
import numpy as np
import pandas as pd
from preamble import *
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
rnd = np.random.RandomState(0)
X_org = rnd.normal(size=(1000, 3))
w = rnd.normal(size=3)

X = rnd.poisson(10 * np.exp(X_org))
y = np.dot(X_org, w)
```

```python
print("Number of feature appearances:\n{}".format(
    np.bincount(X[:, 0])))
```

```python
bins = np.bincount(X[:, 0])
plt.bar(range(len(bins)), bins, color='b')
plt.ylabel("Number of appearances")
plt.xlabel("Value")
```

```python
from sklearn.linear_model import Ridge
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=0)
ridge = Ridge().fit(X_train, y_train)
score = ridge.score(X_test, y_test)
print("Test score: {:.3f}".format(score))
```

```python
X_train_log = np.log(X_train + 1)
X_test_log = np.log(X_test + 1)
```

```python
plt.hist(np.log(X_train_log[:, 0] + 1), bins=25, color='gray')
plt.hist(np.log(X_train_log[:, 1] + 1), bins=25, color='red')
plt.hist(np.log(X_train[:, 1] + 1), bins=25, color='blue')
plt.ylabel=("Number of appearances")
plt.xlabel("Value")
```

```python
score = Ridge().fit(X_train_log, y_train).score(X_test_log, y_test)
print("Test score: {:.3f}".format(score))
```
